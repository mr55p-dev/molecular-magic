```python
import numpy as np
import pandas as pd
X = np.load("static_data/create_features_output/data/features.npy")
```

```python
frame = pd.DataFrame(X)

def createform(x) -> str:
    # return 0
    x_floor = (x / 100).astype('int')
    return f"C{x_floor[0]}O{x_floor[1]}H{x_floor[2]}N{x_floor[3]}"

frame["formulae"] = np.apply_along_axis(createform, 1, X[:, -15:-11])
```

```python
frame["formulae"].nunique()
```

```python
frame.groupby(["formulae"]).count()[0].plot().hist("0")
```
